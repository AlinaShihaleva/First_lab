```python
# 1. Подключение Pandas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# 2. Создание таблицы с помощью кода
url = 'https://conomy.ru/education/articles/147'

id_table = 0

df_power = pd.read_html(url, header = 0, index_col = 0)[id_table]
df_power
```

```python
# 3. Сортировка
df_power = df_power.sort_values(['«Интер РАО»'])
df_power
```

```python
# 4. Добавление столбца
df_power['Итого'] = df_power['«Интер РАО»'] + df_power['«РусГидро»'] + df_power['«Юнипро»'] + df_power['«Мосэнерго»'] + df_power['ОГК-2'] + df_power['ТГК-1'] + df_power['«Энел Россия»']
df_power
```

```python
# 5. Выбор фрагмента таблицы по условию
# Выбрать типы генерации компании «РусГидро» с мощностью меньше 1000 МВт
print('Типы генерации компании «РусГидро» с мощностью меньше 1000 МВт | Мощность')
print([df_power['«РусГидро»'][(df_power['«РусГидро»'] < 1000.)]])

# Выбрать типы генерации компании ТГК-1 с мощностью больше 3000 МВт
print('\nТипы генерации компании ТГК-1 с мощностью больше 3000 МВт | Мощность')
print(df_power['ТГК-1'][(df_power['ТГК-1'] > 3000.)])

# Выбрать типы генерации компании с мощностью выше 1000 МВт
print('\nТипы генераций с мощностью выше 10000 МВт | Итого')
print(df_power.Итого[df_power.Итого > 10000])
```

```python
# 6. Выборка нужных фрагментов
# Через iloc по номерам строк и столбцов, номера идут с нуля
print('Данные из первой строки и второго столбца')
print(df_power.iloc[0 , 1])
print()
print('Данные из всех строк, кроме двух последних, и из столбцов с первого по четвертый включительно')
print(df_power.iloc[: -2, 0 : 3])
print()
print('Данные из нечетных строк и третьего столбца')
print(df_power.iloc[ : : 2, 3])
print()
```

```python
# Переименование столбцов
df_power_2 = pd.read_html(url, header = 0, index_col = None)[id_table]
df_power_2['Итого'] = df_power_2['«Интер РАО»'] + df_power_2['«РусГидро»'] + df_power_2['«Юнипро»'] + df_power_2['«Мосэнерго»'] + df_power_2['ОГК-2'] + df_power_2['ТГК-1'] + df_power_2['«Энел Россия»']
df_power_2.columns = ['Типы генерации', '«Интер РАО»', '«РусГидро»', '«Юнипро»',	'«Мосэнерго»',	'ОГК-2',	'ТГК-1',	'«Энел Россия»', 'Итого']
df_power_2
```

```python
# 7. Среднестатистические расчеты по сгруппированным данным
# Суммарная мощность по типам генераций
print(df_power_2.groupby(['Типы генерации'])['Итого'].sum())
print()

input_matrix = df_power_2.values # перевод таблицы в матрицу

# Можно посчитать среднее, минимальное, максимальное и прочие показатели
print('Максимальные значения мощности каждой компании: ')
for i in range(1, len(df_power_2.columns)):
  print(df_power_2.columns[i], input_matrix[: , i].max())
```

```python
# 8. Гистограмма
axes = df_power.iloc[ : , :7].plot(kind = 'bar', figsize=(21, 5), subplots = False)

# Добавление точных значений каждому столбцу
for p in axes.patches:
    axes.annotate(str(p.get_height()), (p.get_x(), p.get_height() * 1.01))

axes.grid(True)
```

```python
# 9. Круговая диаграмма
input_matrix = df_power.values # перевод таблицы в матрицу

# Получить сумму значений по столбцам, отвечающим за разные виды компаний
generation_sum = (input_matrix[: , : 7]).sum(axis = 0)

# Pie chart, where the slices will be ordered and plotted counter-clockwise:
labels = df_power.columns[ : 7]

fig1, ax1 = plt.subplots(figsize = (10, 10))
ax1.pie(generation_sum, labels=labels, autopct='%1.1f%%')
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
